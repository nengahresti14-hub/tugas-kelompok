```python
class AntrianTiket:
    def __init__(self, nama_event, kuota_antrian):
        # Inisialisasi Array (List) untuk menampung antrian
        self.antrian = []
        self.nama_event = nama_event
        self.maksimal = kuota_antrian

    def enqueue(self, nama_user):
        """Menambahkan user ke dalam antrian (INPUT Nama User)"""
        # Cek apakah kuota penuh?
        if len(self.antrian) < self.maksimal:
            self.antrian.append(nama_user)
            print(f"BERHASIL: {nama_user} masuk antrian {self.nama_event}. Posisi: {len(self.antrian)}")
        else:
            # Output jika YA (Kuota Penuh)
            print(f"GAGAL: Antrian Penuh! Maaf {nama_user}, kuota {self.nama_event} sudah mencapai batas.")

    def dequeue(self):
        """Memproses/mengambil orang terdepan dari antrian (PROSES Dequeue)"""
        # Cek apakah antrian kosong sebelum memproses
        if not self.is_empty():
            # Menggunakan pop(0) untuk mengambil data paling depan (FIFO)
            pembeli = self.antrian.pop(0)
            print(f"PROSES: 1 Tiket Terbit untuk: {pembeli}!")
            return pembeli
        else:
            print("GAGAL: Tidak ada antrian yang bisa diproses.")
            return None

    def peek(self):
        """Melihat siapa yang berada di urutan paling depan"""
        if not self.is_empty():
            print(f"INFO: Orang terdepan saat ini: {self.antrian[0]}")
            return self.antrian[0]
        else:
            print("INFO: Antrian masih kosong.")
            return None

    def display(self):
        """Tampilkan seluruh posisi antrian"""
        if self.is_empty():
            print("\nSTATUS: Antrian Sedang Kosong.")
        else:
            print(f"\nDAFTAR ANTRIAN {self.nama_event.upper()}:")
            for i, user in enumerate(self.antrian):
                print(f"   {i+1}. {user}")
            print(f"   Total: {len(self.antrian)} orang di dalam barisan.\n")

    def is_empty(self):
        """Fungsi pembantu untuk cek status kosong"""
        return len(self.antrian) == 0



war_tiket = AntrianTiket("Konser Dewa 19", 3)

# 2. ENQUEUE: Input Nama User
war_tiket.enqueue("Andi")
war_tiket.enqueue("Budi")
war_tiket.enqueue("Cici")
war_tiket.enqueue("Deni")

# 3. DISPLAY: Tampilkan Posisi
war_tiket.display()

# 4. PEEK: Lihat urutan depan
war_tiket.peek()

# 5. DEQUEUE: Proses Bayar / Tiket Terbit
war_tiket.dequeue()

# 6. Lihat sisa antrian setelah diproses
war_tiket.display()
```
